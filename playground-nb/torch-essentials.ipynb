```python
# Basic Operations
tensor1 = torch.tensor([3,-5,7])
add = tensor1 + 10  # OR torch.add(tensor1,10)
sub = tensor1 - 10  # OR torch.sub(tensor1,10)
mul = tensor1 * 4   # OR torch.mul(tensor1, 4)
print(tensor1)
print(add)
print(sub)
print(mul)
print(torch.matmul(tensor1, tensor1))
```

```python
# Shapes need to be in the right way  
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)

tensor_B = torch.tensor([[7, 10],
                         [8, 11], 
                         [9, 12]], dtype=torch.float32)

torch.matmul(tensor_A, tensor_B.T) # Changed tensor_B to transpose, thus matrix multiplication is possible
```

```python
torch.manual_seed(149)  # Manual seeds basically help in order of random generation of values
tens_rand = torch.rand(4,2)
torch.manual_seed(149)
tens_rand2 = torch.rand(4,2)
tens_rand == tens_rand2
```

```python
x = torch.arange(1,8)
# Resizing: Add an extra dimension
x_reshaped = x.reshape(1, 7)
x_reshaped, x_reshaped.shape

# Change view (keeps same data as original but changes view)
z = x.view(1, 7)
z, z.shape  # Same as x_reshaped, but only a view of x

# Changing z changes x
z[:, 0] = 5
z, x    # z = [[5,2,3,4,5,6,7]]
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=1) # dim 0 does on first dimension, and so on
x_stacked
```

```python
print(f"Previous tensor: {x_stacked}")
print(f"Previous shape: {x_stacked.shape}")

# Add extra dimension to x_stacked
x_squeezed = x_stacked.unsqueeze(dim=0)
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
# Create tensor with specific shape
x_original = torch.rand(size=(224, 224, 3))

# Permute the original tensor to rearrange the axis order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
```

```python
# Set device type, cuda, cpu, mps
if torch.cuda.is_available():
    device = "cuda" # Use NVIDIA GPU (if available)
elif torch.backends.mps.is_available():
    device = "mps" # Use Apple Silicon GPU (if available)
else:
    device = "cpu" # Default to CPU if no GPU is available
device
```

```python
# Create tensor (default on CPU)
tensor = torch.tensor([1, 2, 3])

# Tensor not on GPU
print(tensor, tensor.device)

# Move tensor to GPU (if available)
tensor_on_gpu = tensor.to(device)
tensor_on_gpu
```
